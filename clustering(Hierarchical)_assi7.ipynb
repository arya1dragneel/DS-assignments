Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop - urban population in different places of United States
Rape - Rape rate in different places of United States

```python
# Hierarchical clustering
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime_data=pd.read_csv("/content/crime_data.csv")
crime_data
```

```python
#rename
crime_data1=crime_data.rename({'Unnamed: 0':'State'},axis=1)
crime_data1.head()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(crime_data1.State,crime_data1.Murder,color='r')
plt.scatter(crime_data1.State,crime_data1.Assault,color='g')
plt.scatter(crime_data1.State,crime_data1.UrbanPop,color='b')
plt.scatter(crime_data1.State,crime_data1.Rape,color='y')
plt.xlabel('state')
plt.ylabel('rate')
plt.show()
```

```python
#Data normalisation
#Defining the normaliation function
def normalf(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
#normalize the dataframe
crime_data_normal=normalf(crime_data1.iloc[:,1:])
```

```python
crime_data_normal
```

```python
# create a dendogram
#plotting the pendogram-eard method
plt.figure(figsize=(12,6))
dendogram=sch.dendrogram(sch.linkage(crime_data_normal,method='ward'))
```

```python
#fit and predict
#specifying hierarchical clustering ,clusters as 3(we are considering n of clusteres=3)
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
#prediction - there is no centroid in agglomerative options
y_hc=hc.fit_predict(crime_data_normal)
y_hc
```

```python
#clusters are formed
# creating new dataframe column are clusters
clusters=pd.DataFrame(y_hc,columns=['clusters'])
crime_data1['Hierarchical_clusterid']=clusters
```

```python
crime_data1
```

```python
crime_data1.groupby('Hierarchical_clusterid').agg(['mean']).reset_index()
```

```python
crime_data1['Hierarchical_clusterid'].value_counts()
```

```python
#plot clusters
plt.figure(figsize=(10,7))
plt.scatter(crime_data1['Hierarchical_clusterid'],crime_data1['Murder'],c=hc.labels_)
```
