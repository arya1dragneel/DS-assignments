```python
#rename of the column
#renaming column1 to state
crime_data=df.rename({'Unnamed: 0':'State'},axis=1)
crime_data
```

```python
#normalisation function
scaler=StandardScaler()
scaled_crime_data=scaler.fit_transform(crime_data.iloc[:,1:])
```

```python
scaled_crime_data
```

```python
#identify the best k-value using the elbow method
wcss=[]
for i in range(1,11):
  kmeans =KMeans(n_clusters=i,random_state=0)
  kmeans.fit(scaled_crime_data)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#k=4 is the best we can see
#build cluster algorithm
clusters_new =KMeans(4,random_state=42)  #randomly divide the data into 4 partitions
clusters_new
```

```python
clusters_new.fit(scaled_crime_data)
clusters_new.labels_
```

```python
crime_data['ClusterID_new']=clusters_new.labels_
crime_data.head()
```

```python
#Average values of all data points within one clustersx for profiling purposes
clusters_new.cluster_centers_
```
