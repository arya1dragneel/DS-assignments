```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='darkgrid', palette='viridis')
%matplotlib inline
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
df= pd.read_csv('/content/book (1).csv', encoding='latin-1')
df
```

```python
#EDA
```

```python
df.rename(columns={"User.ID":"user_id","Book.Title":"book_title","Book.Rating":"book_rating"},inplace=True)
```

```python
df1=df.iloc[:,1:]
df1
```

```python
# Sort by User IDs
df1.sort_values(['user_id'])
```

```python
# number of unique users in the dataset
len(df1['user_id'].unique())
```

```python
# number of unique books in the dataset
len(df1['book_title'].unique())
```

```python
# converting long data into wide data using pivot tabl
df2=df1.pivot_table(index='user_id',columns='book_title',values='book_rating').reset_index(drop=True)
df2
```

```python
# Replacing the index values by unique user Ids
df2.index=df1['user_id'].unique()
df2
```

```python
# Impute those NaNs with 0 values
df2.fillna(0,inplace=True)
df2
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(df2.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids
user_sim2.index=df1['user_id'].unique()
user_sim2.columns=df1['user_id'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)
```
