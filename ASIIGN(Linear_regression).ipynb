# Model building

```python
#import statsmodel
import statsmodels.formula.api as smf
```

```python
model = smf.ols('SAL~YE',data=salary_df).fit()
```

```python
model = smf.ols('DT~ST',data=delivery_df).fit()
```

```python
#Model parameters
model.params
```

```python
#Model parameters
model.params
```
